1. Bir Python programı yazın ve bu programı çalıştırdığınızda bir metin dosyası oluşturmasını sağlayın. 

2.  Bu metin dosyası, öğrenci bilgilerini (ad, soyad, öğrenci numarası) saklamalıdır. 

3.  Kullanıcıya, yeni öğrenci bilgileri eklemek veya mevcut bilgilere erişmek için seçenekler sunun. 

4.  Yeni öğrenci bilgileri eklerken, kullanıcıdan ad, soyad ve öğrenci numarası girmesini isteyin ve bu bilgileri metin dosyasına ekleyin. 

5.  Mevcut bilgilere erişirken, kullanıcıdan bir öğrenci numarası girmesini isteyin ve bu öğrenci numarasına sahip öğrencinin bilgilerini ekranda gösterin. 

6.  Kullanıcı işlemi bitirince dosyayı kapatmayı unutmayın. 

```python
dosya = input("dosya ismini girin veya yeni dosya olusturun:")
with open(dosya, "r", encoding = "utf-8") as file:
    pass

while True:
    
    x = input("mevcut bilgilere erismek icin m, yeni bilgi eklemek icin y, cikmak icin q tusuna basiniz:")
    
    if x.lower() == 'm': # her satiri kontrol edecek ve o sayi varsa o satiri dondurecek, yoksa boyle bir kayit bulunamadi ciktisi verilecek
        
        with open(dosya, "r", encoding="utf-8") as file:
                file.read()
                file.seek(0)  # Dosyanın okuma imlecini dosyanın başına getiriyoruz
                
                while True:
                    
                    y = input("Bilgisine erişmek istediğiniz öğrencinin numarasini giriniz, cikmak icin q tusuna basiniz: ")
                    found = False
                    for line in file.readlines():
                        if y in line:
                            print(line)
                            found = True
                    file.seek(0)
                    if y == 'q':
                        break
                    if not found:
                        print(f"{y} öğrenci numarali bir kayit bulunamadi.")
                        
    elif x.lower() == 'y':
        
        while True:
            ogrbilgi = input("isim, soyisim, ve ogrenci numarasi bilgilerini girniz, orn: Ayse Deniz, 32. (cikmak icin q a basiniz)")
            
            if ogrbilgi == 'q':
                break
            
            with open(dosya, "a") as file:
                file.write(f"{ogrbilgi}\n")
                print(f"{ogrbilgi} dosyaya eklendi")
            
        with open(dosya, "r") as file:
            print("\nveri girisi yaptiktan sonra dosya icerigi:\n",file.read())
            
    elif x.lower() == 'q':
        break
```

```python
# Hocanin cozumu

# Dosya adı
dosya_adi = "ogrenci_veritabani.txt"

# Veritabanı dosyasını oluştur
def veritabani_olustur():
    with open(dosya_adi, "w") as dosya:
        dosya.write("Ad,Soyad,OgrenciNo\n")

# Yeni öğrenci bilgisi ekleme
def ogrenci_ekle(ad, soyad, ogrenci_no):
    with open(dosya_adi, "a") as dosya:
        dosya.write(f"{ad},{soyad},{ogrenci_no}\n")

# Öğrenci bilgilerini gösterme
def ogrenci_goster(ogrenci_no):
    with open(dosya_adi, "r") as dosya:
        for satir in dosya.readlines():
            veriler = satir.strip().split(",")
            if veriler[2] == ogrenci_no:
                return f"Ad: {veriler[0]}, Soyad: {veriler[1]}, Ogrenci No: {veriler[2]}"
        return "Öğrenci bulunamadı."

# Ana program
veritabani_olustur()
while True:
    print("1. Yeni öğrenci ekle")
    print("2. Öğrenci bilgilerini göster")
    print("3. Çıkış")
    secim = input("Seçiminizi yapın: ")
    
    if secim == "1":
        ad = input("Öğrenci Adı: ")
        soyad = input("Öğrenci Soyadı: ")
        ogrenci_no = input("Öğrenci No: ")
        ogrenci_ekle(ad, soyad, ogrenci_no)
    elif secim == "2":
        ogrenci_no = input("Öğrenci No girin: ")
        bilgi = ogrenci_goster(ogrenci_no)
        print(bilgi)
    elif secim == "3":
        break
    else:
        print("Geçersiz seçenek, tekrar deneyin.")
```
