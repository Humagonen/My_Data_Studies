```python
m = np.linspace(0, 10, 11) 
c = 3 * 10**8 
E = m * c**2  
```

```python
plt.plot(m, E, color= 'green')
plt.title('E= mc^2')
plt.ylabel('E in J')
plt.xlabel('M in g')

plt.yscale("log")                    # y-eksenini logaritmik ölçeğe ayarlama
plt.grid(which="both", axis="y")     # y-ekseninde ızgara çizgilerini ayarlama

plt.show()
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapalim

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.plot(labels, july16_2007, color = 'tab:blue', label = 'july16_2007');
plt.plot(labels, july16_2020, 'orange', label = 'july16_2020');

plt.legend() 
plt.show()
```

```python
plt.plot(labels, july16_2007, color = 'tab:blue', label = 'july16_2007');
plt.plot(labels, july16_2020, 'orange', label = 'july16_2020');

plt.legend(bbox_to_anchor = (1.3, 0.56))
plt.show()
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
fig, ax = plt.subplots(2, 1) # 2 row, 1 col olustur 

ax[0].plot(labels, july16_2007, color = 'tab:blue') # 0. row' a yerlestir
ax[0].set_title("july16_2007") # title
ax[1].plot(labels, july16_2020, color = 'tab:blue') # 1. row' a yerlestir
ax[1].set_title("july16_2020") # title

plt.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize= (8,6))

color = 'blue'
ax1.set_ylabel('July16_2007', color=color, size= 14)
ax1.plot(labels, july16_2007, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Creating a second y-axis on the right
ax2 = ax1.twinx()

color = 'red'
ax2.set_ylabel('July16_2020', color=color, size= 14)
ax2.plot(labels, july16_2020, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.show()
```
